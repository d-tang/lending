````python
def create_target(default_list, df):
    # Creates target feature - '1' if default, '0' if not
    # Takes dataframe as well as list of loan statuses to include as 'default'
    df['target'] = [1 if n in default_list else 0 for n in df['loan_status']]
````

````python
df1 = create_target(['In Grace Period', 'Charged Off', 'Default', 'Late (31-120 days)', 'Late (16-30 days)'], df)
````

````python
# function to extract month into numeric

def month_numeric(df, col_name):
    month_dict = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
                 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
    df['{}'.format(col_name)+'_num'] = df[col_name].apply(lambda x: month_dict[str(x).split('-')[0]])
````
